```python
import matplotlib.pyplot as plt
import cv2
import os

gray_path = 'C:\\Users\\espebh\\Documents\\Thesis\\data\\semiautomatically generated data\labled data\\tank 9 data\\dot_dataset_1000\\images_gray'
lap_path = 'C:\\Users\\espebh\\Documents\\Thesis\\data\\semiautomatically generated data\labled data\\tank 9 data\\dot_dataset_1000\\images_lap'
mask_path = 'C:\\Users\\espebh\\Documents\\Thesis\\data\\semiautomatically generated data\labled data\\tank 9 data\\dot_dataset_1000\\images_mask'
img_path = 'C:\\Users\\espebh\\Documents\\Thesis\\data\\semiautomatically generated data\labled data\\tank 9 data\\augmented_data_1000\\images'

ex_img = '5.14'
fig, ax = plt.subplots(2,2,figsize = (15,10))
img_gray = cv2.imread(os.path.join(gray_path, ex_img + '.jpg'))
img_lap = cv2.imread(os.path.join(lap_path, ex_img + '.jpg'))
img_thresh = cv2.imread(os.path.join(mask_path, ex_img + '.jpg'))
img_full = cv2.imread(os.path.join(img_path, ex_img + '.jpg'))

ax[0][0].imshow(img_full)
ax[0][0].set_title('Color image', fontsize = 25)
ax[0][0].set_xticks([])
ax[0][0].set_yticks([])
ax[0][1].imshow(img_gray, cmap = 'gray')
ax[0][1].set_title('Gray scale image', fontsize = 25)
ax[0][1].set_xticks([])
ax[0][1].set_yticks([])
ax[1][0].imshow(img_lap)
ax[1][0].set_title('Image Laplacian', fontsize = 25)
ax[1][0].set_xticks([])
ax[1][0].set_yticks([])
ax[1][1].imshow(img_thresh)
ax[1][1].set_title('Adaptive thresholding image', fontsize = 25)
ax[1][1].set_xticks([])
ax[1][1].set_yticks([])
```
